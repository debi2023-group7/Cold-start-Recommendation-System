# Problem Statement
-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Cold Start Problem in Recommendation Systems**

The cold start problem in recommendation systems refers to the challenging situation where the system encounters new users or items for which it lacks sufficient historical data to generate accurate and personalized recommendations. This issue arises when the system is in its early stages or when it expands to include new user segments or items that have limited or no interaction history.

In the context of user cold start, the recommendation system struggles to provide relevant suggestions to new users who have just joined the platform, as it lacks insights into their preferences, behaviors, and interests. Since the system relies on historical data to understand user preferences, it becomes difficult to provide personalized recommendations to users with little or no prior interactions.

Similarly, item cold start poses a challenge when new products or items are introduced to the recommendation system. Without historical usage data or user feedback on these items, the system cannot accurately gauge their attributes and how well they match the preferences of users. This leads to difficulties in presenting personalized recommendations for these new items.

The cold start problem is particularly critical because it impacts the overall user experience and system performance. In the absence of historical data, the recommendation system might resort to generic or popular recommendations, leading to poor user satisfaction and reduced engagement.

Addressing the cold start problem requires innovative solutions that can effectively handle scenarios with limited or no user-item interaction data. This can involve leveraging auxiliary information, such as user demographics, social connections, or content attributes of items, to make educated inferences and create initial recommendations. Furthermore, actively engaging new users to gather preferences through explicit feedback or interactions can aid in mitigating the user cold start problem. For item cold start, employing content-based filtering, collaborative filtering with item attributes, or hybrid approaches can assist in offering relevant recommendations for new items.

For our case we are working only with limited or no user-item interaction data.

# DataSets

**Tenrec dataset**

The Tenrec dataset suite is a collection of user behavior logs from Tencent's recommendation platforms, QQ BOW (QB) and QQ KAN (QK). The dataset includes user feedback from four different scenarios: QK-video, QK-article, QB-video, and QB-article. In the QK/QB platforms, items can be either news articles or videos.

**Data Collection**:

- The user behavior logs were collected from September 17 to December 07, 2021.
- Around 5.02 million users were randomly selected from the QK-video database, with each user having at least 5 video clicking behaviors.
- Feedback data consists of positive feedback, such as video clicks, shares, likes, and follows, as well as negative feedback, where there was exposure but no user action.
- The dataset contains around 142 million video clicks, 10 million likes, 1 million shares, and 0.86 million follows, alongside 3.75 million videos.
- Age and gender features are available for users, and video type features are provided for items.

**Data Distribution**:

- The item popularity follows a long-tail distribution, which is common in recommendation systems.
- The session length distribution shows that the number of sessions with a length in the range [0, 20] accounts for 53% of all sessions.

**Data Overlapping**:

- There is overlap in users and items across the four scenarios.
- For overlapped users, there are 268,207 users shared between QK-video and QK-article, 3,261 between QK-video and QB-video, and 58 between QK-video and QB-article.
- For overlapped items, there are 78,482 videos shared between QK-video and QB-video.

Overlapped users and items can be associated using their unique IDs, making the Tenrec dataset suitable for studying the TF (Transfer Learning) and CDR (Cross-Domain Recommendation) tasks.

# Model

**BERT4Rec**

BERT4Rec is a state-of-the-art recommendation model that adapts the BERT (Bidirectional Encoder Representations from Transformers) architecture, originally designed for natural language processing tasks, to address the recommendation problem. BERT4Rec leverages the power of pre-training on large-scale text data and fine-tuning on recommendation datasets to generate highly effective and context-aware recommendations.

**Description of BERT4Rec**:

- **BERT Architecture**: BERT4Rec is built upon the BERT architecture, which is a transformer-based neural network model designed to capture bidirectional contextual information from text. BERT's attention mechanism allows it to efficiently process and represent long-range dependencies in sequential data, making it ideal for capturing complex patterns in user-item interactions.

- **Fine-tuning for Recommendations**: BERT is fine-tuned on a specific recommendation dataset. The recommendation dataset typically consists of user-item interaction data, such as user-item click histories or ratings. The objective of fine-tuning is to adapt the pre-trained BERT model to better understand the sequential patterns and user preferences specific to the recommendation domain.

- **Sequential Modeling**: BERT4Rec excels in handling sequential data, which is common in recommendation systems. It can take into account the temporal order of user interactions with items and learn representations that capture the evolving preferences of users over time.

- **Contextual User and Item Representations**: BERT4Rec generates contextual embeddings for both users and items. By considering the entire interaction history of users and capturing the semantics of items, the model can create context-aware representations that reflect the dynamic nature of user preferences.

- **Personalized Recommendations**: BERT4Rec can generate personalized recommendations by leveraging the contextual embeddings of users and items. This allows the model to tailor its recommendations based on the unique preferences and interests of individual users.

- **State-of-the-Art Performance**: BERT4Rec has achieved impressive results in various recommendation tasks and benchmarks. Its ability to capture complex interactions and effectively handle sequential data has made it a popular choice among researchers and practitioners in the recommendation community.

- **Computational Complexity**: While BERT4Rec offers exceptional performance, it comes with increased computational requirements compared to traditional recommendation algorithms. Pre-training and fine-tuning on large-scale datasets demand substantial computational resources and training time.

# Evaluation Metrics 
-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Recall@k**

Recall@k is a performance metric commonly used in information retrieval and recommendation systems. It measures the proportion of relevant items that are successfully retrieved among the top k items recommended to a user. In other words, Recall@k evaluates how well a recommendation model can capture the items that a user is likely to find relevant within the first k recommendations.

A high Recall@k value indicates that the model is effective in identifying and presenting relevant items to users, making it more likely that users will find items they are interested in. On the other hand, a low Recall@k value suggests that the model is missing out on many relevant items, potentially leading to suboptimal user experiences.

Recall@k is just one of several evaluation metrics used to assess recommendation systems.

------------------------------------------------------
**NDCG@k**

Normalized Discounted Cumulative Gain (NDCG@k) is an evaluation metric commonly used to assess the performance of recommendation systems and information retrieval models. It measures the quality of the ranked list of recommended items by taking into account both relevance and position in the list.

NDCG@k calculates the relevance of each recommended item based on a graded relevance scale (e.g., 0, 1, 2, 3), where higher values indicate higher relevance. The graded relevance is then discounted based on the item's position in the list, giving higher weight to items that appear at the top. This normalization ensures that the metric takes into account both the relevance of recommended items and the diversity of the list.

A perfect NDCG@k score of 1.0 indicates that the model has recommended the most relevant items at the top, providing the user with highly useful recommendations. Conversely, a lower NDCG@k score suggests that the model's recommendations are less relevant or not well-ordered, potentially leading to a less satisfactory user experience.

NDCG@k is just one of several evaluation metrics used in recommendation systems and information retrieval tasks

```python
def set_seed(seed, re=True):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed) # if you are using multi-GPU.
    if re:
        torch.backends.cudnn.benchmark = False
        torch.backends.cudnn.deterministic = True
    else:
        torch.backends.cudnn.benchmark = True
        torch.backends.cudnn.deterministic = False
```
